```python
import numpy as np
def AND (x1, x2):
  x = np.array([x1,x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  temp = np.sum(w * x) + b
  if temp <=0 :
    return 0
  else:
    return 1

def OR(x1, x2):
  x = np.array([x1,x2])
  w = np.array([1.0,1.0])
  b = -0.5
  temp = np.sum(w * x) + b
  if temp <= 0:
    return 0;
  else:
   return 1;

def NAND (x1, x2):
  x = np.array([x1,x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  temp = np.sum(w * x) + b
  if temp <=0 :
    return 1
  else :
    return 0

def XNOR(x1, x2):
  and_res = AND(x1, x2)
  not_or_res = NAND(OR(x1, x2), OR(x1, x2))
  return OR(and_res, not_or_res)

print(XNOR(0,0))
print(XNOR(0,1))
print(XNOR(1,0))
print(XNOR(1,1))
```
