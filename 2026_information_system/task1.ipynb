# 課題1: matplotlib によるグラフ描画

授業で扱った `np.sin()` / `np.cos()` 以外の関数をプロットする。

```python
import numpy as np
import matplotlib.pyplot as plt
```

## プロットする関数

今回は以下の3つを選んだ。

| 関数 | 式 |
|---|---|
| 指数関数 | $y = e^x$ |
| 対数関数 | $y = \ln(x)$ |
| タンジェント | $y = \tan(x)$ |

```python
# --- 指数関数と対数関数 ---
x_exp = np.linspace(-2, 3, 300)
x_log = np.linspace(0.01, 5, 300)   # log は x > 0

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# (1) 指数関数
axes[0].plot(x_exp, np.exp(x_exp), color='tab:blue', lw=2)
axes[0].set_title(r'$y = e^x$', fontsize=14)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_ylim(-1, 20)
axes[0].axhline(0, color='k', lw=0.8, ls='--')
axes[0].axvline(0, color='k', lw=0.8, ls='--')
axes[0].grid(True, alpha=0.3)

# (2) 対数関数
axes[1].plot(x_log, np.log(x_log), color='tab:orange', lw=2)
axes[1].set_title(r'$y = \ln(x)$', fontsize=14)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].axhline(0, color='k', lw=0.8, ls='--')
axes[1].axvline(0, color='k', lw=0.8, ls='--')
axes[1].grid(True, alpha=0.3)

# (3) タンジェント（不連続点付近は除外）
x_tan = np.linspace(-np.pi/2 + 0.05, np.pi/2 - 0.05, 300)
axes[2].plot(x_tan, np.tan(x_tan), color='tab:green', lw=2)
axes[2].set_title(r'$y = \tan(x)$', fontsize=14)
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].set_ylim(-10, 10)
axes[2].axhline(0, color='k', lw=0.8, ls='--')
axes[2].axvline(0, color='k', lw=0.8, ls='--')
axes[2].grid(True, alpha=0.3)

plt.suptitle('授業で扱った sin/cos 以外の関数プロット', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()
```

## 各関数について

- **指数関数 $e^x$**：x が増えるにつれて急激に増加する。x=0 のとき y=1。
- **対数関数 $\ln(x)$**：指数関数の逆関数。x=1 のとき y=0 で，x が大きくなるほど緩やかに増加する。x≤0 では定義されない。
- **タンジェント $\tan(x)$**：$x = \pm\pi/2$ で不連続（発散）になる。$\sin(x)/\cos(x)$ と等しい。
