#### クラス
int や str などのデータ型は「組み込み」のデータ型と呼ばれ，はじめからPythonに組み込まれたデータ型である．新しいクラスを定義することで，ユーザが自分で独自のデータ型を作成する事ができる．クラスにはメソッド (クラス用の関数) や属性 (変数) を定義できる．

##### クラスのフォーマット
```
class クラス名:
    def __init__(self, 引数, …): # コンストラクタ
    …
    def メソッド名1(self, 引数, …): # メソッド1
    …
    def メソッド名2(self, 引数, …): # メソッド2
    …
```

`__init__` は特別なメソッドで，初期化の役割がある．この初期化用のメソッドはコンストラクタと呼ばれ，クラスのインスタンス (クラスが利用できる状態になったもの) が作成される際に一度だけ実行される．

以下で簡単なクラスを作成してみる．

````python
class Man:
    def __init__(self, name): # コンストラクタ(「_」を前後に2個ずつつける)
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello " + self.name + "!")

    def goodbye(self,name):
        print("Good-bye " + name + "!")

m = Man("Ikeuchi") # インスタンスを作る
m.hello()
m.goodbye("Chiba")
````

````python
class Man2:
    def __init__(self, name): # コンストラクタ(「_」を前後に2個ずつつける)
        self.name = name
        self.count = 0
        print("Initialized!")

    def hello(self):
        print("Hello " + self.name + "!")
        self.count += 1
        print('count = ' + str(self.count))

    def goodbye(self,name):
        print("Good-bye " + name + "!")

m = Man2("Yamada") # インスタンスを作る
m.hello()
m.hello()
m.hello()
````

ここでは Man という新しいクラスを定義している．この Man というクラスから，m というインスタンスを生成している．Man クラスのコンストラクタは name という引数をとり，その引数でインスタンス変数である `self.name` を初期化する．インスタンス変数とは，個々のインスタンスに格納される変数のことで，Python では `self.name`のように，self のあとに属性名を書くことでインスタンス変数の作成，及び，アクセスができる．

> ちょっとややこしいですよね．
> 関数は処理をひとまとめにして，何回も再利用できるようにしましたね．
> クラスはなにが違うかというと，変数 (上ではself.nameですね) を持っていて，それを覚えておける(インスタンスをつくって，その中に保存しておける) というところです．
> 例えば，1回目の処理のあと，2回目の処理をやります，というときに，1回目の処理結果を覚えておかないと2回目の処理ができないということがありますよね．
> そういう時は関数でやるやり方もありますが，クラスを使うと，1回目の処理結果をインスタンスの中に保存しておけるので，2回目の処理ができます．
> ほかにもクラスを使う理由はあるのですが，ひとまずこの辺りはまあそんなもんかな，ということで良いです．
> 今後，学習を進めていくと使う機会がありますので，「そういえばそんなこともやったな」ということで思い出してください．

先程のグラフにcos 関数も描画する．また，タイトルやx軸ラベル名の描画など，pyplot の他の機能も使う．

````python
y1 = np.sin(x)
y2 = np.cos(x)

````

````python
# グラフの描画
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle = "--", label = "cos") # 破線で描画
plt.xlabel("x") # x軸のラベル
plt.ylabel("y") # y軸のラベル
plt.title('sin & cos') # タイトル
plt.legend() # 凡例表示
plt.show()
````
