```python
import numpy as np

def _numerical_gradient_no_batch(f, x):
    h = 1e-4
    grad = np.zeros_like(x)

    for idx in range(x.size):
        tmp_val = x[idx]

        x[idx] = tmp_val + h
        fxh1 = f(x)

        x[idx] = tmp_val - h
        fxh2 = f(x)

        grad[idx] = (fxh1 - fxh2) / (2 * h)
        x[idx] = tmp_val

    return grad
```

```python
def function_3d(variables):
    x = variables[0]
    y = variables[1]
    z = variables[2]
    return x**2 * np.sin(y) + np.exp(-z) * np.cos(x * y)
```

```python
variables = np.array([1.0, np.pi / 4, 0.5])
print(_numerical_gradient_no_batch(function_3d, variables))
```
