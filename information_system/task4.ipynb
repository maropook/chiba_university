# 課題
ORの関数をステップ関数と行列の積を使って書き換えてください．
なお，ANDの関数は以下のようにして書き換えることができますので，こちらを参考にしてください．

````python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
````

上記は書き換える前のANDの関数です．

````python
def step_function(x):
    return np.array(x > 0, dtype=int)
````

上記はステップ関数です．

````python
def AND2(x):
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.dot(x, w) + b

    return step_function(tmp)
````

上記のように，入力と重みの掛け算と足し算の計算を行列の積```np.dot()```で計算して，その結果にバイアス```b```を加算して，ステップ関数```step_function()```に入力しています．

````python
x = np.array([0 ,0])
x
````

````python
AND2(x)
````

ステップ関数と行列の積を使って書き換えられた```AND2()```に0, 1の組合わせを入力すると，書き換え前の```AND()```と同じ結果になることがわかります．

では，下記のORの関数を同様にしてステップ関数と行列の積を使って書き換えてください．

````python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([1.0, 1.0])
    b = -0.5
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
````

````python
def OR2(x):
    w = np.array([1.0, 1.0])
    b = -0.5
    tmp = np.dot(x, w) + b

    return step_function(tmp)
````

````python
x_00 = np.array([0, 0])
x_01 = np.array([0, 1])
x_10 = np.array([1, 0])
x_11 = np.array([1, 1])
print(f"OR2(0,0): {OR2(x_00)}")
print(f"OR2(0,1): {OR2(x_01)}")
print(f"OR2(1,0): {OR2(x_10)}")
print(f"OR2(1,1): {OR2(x_11)}")
````
