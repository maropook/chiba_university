```python
import numpy as np
import matplotlib.pyplot as plt

x_exp = np.linspace(-2, 3, 300)
x_log = np.linspace(0.01, 5, 300)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(x_exp, np.exp(x_exp), color='tab:blue', lw=2)
axes[0].set_title(r'$y = e^x$', fontsize=14)
axes[0].set_ylim(-1, 20)
axes[0].grid(True, alpha=0.3)

axes[1].plot(x_log, np.log(x_log), color='tab:orange', lw=2)
axes[1].set_title(r'$y = \ln(x)$', fontsize=14)
axes[1].grid(True, alpha=0.3)

x_tan = np.linspace(-np.pi/2 + 0.05, np.pi/2 - 0.05, 300)
axes[2].plot(x_tan, np.tan(x_tan), color='tab:green', lw=2)
axes[2].set_title(r'$y = \tan(x)$', fontsize=14)
axes[2].set_ylim(-10, 10)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
